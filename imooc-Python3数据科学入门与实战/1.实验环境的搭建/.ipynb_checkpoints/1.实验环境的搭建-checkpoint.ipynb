## Anaconda 和 Jupyter Notebook 介绍

### Conda 的 Environment 管理

* 创建一个新的 environment

    `conda create --name python3 python=3.6`

* 激活一个 environment

    `activate python3  # for Windows`
    
    `source activate python3  # for Linux & Mac`
    
* 退出一个 environment

    `deactivate python3  # for Windows`
    
    `source deactivate python3  # for Linux & Mac`
 
* 删除一个 environment

    `conda remove --name python3 --all`
    
### Conda 的 Package 管理

* Conda 的包管理有点类似 pip

* 安装一个 Python 包

    `conda install numpy`
    
* 查看已安装的 Python 包

    `conda list`
    
    `conda list -n python3  # 查看指定环境安装的 Python 包`
    
* 删除一个 Python 包

    `conda remove -n python3 numpy`
    
### Data Science IDE----------VS -----------Developer IDE

* 以数据为中心-------------------|----------以代码为中心
* 交互式和可视化-----------------|----------类和调试
* 代码比较简单，以script 为主-----|----------代码复杂，以 Program 为主
* 和各种数据源做交互--------------|----------和 git、build、compile 工具交互
* Spyder\Jupyter Notebook\JupyterLab--Pycharm\Visual Studio Code\Sublime Text

### Ipython Kernel

* 主要负责运行用户代码
* 通过 stdin/stdout 和 Ipython shell 交互
* 用 json message 通过 zeroMQ 和 notebook 交互

### Notebook 的文件格式 (.ipynb)

* 由 Ipython Notebook 定义的一种格式 (json )
* 可以读取在线数据，CSV / XLS 文件
* 可以转换成其他格式 (py, html, pdf, md 等)

### NBViewer

* 一个 online 的 ipynb 格式 notebook 展示工具
* 可以通过 URL 分享
* GitHub 集成流 NBViewer
* 通过转换器轻松集成到 Blogs, Emails, Wikis, Books
